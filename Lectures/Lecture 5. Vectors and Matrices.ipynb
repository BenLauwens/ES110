# Vectors in 3D

## Definition

A vector in $\mathbb R^3$ is represented by a column array:
$$\bar{v} = \begin{bmatrix}
v_1\\
v_2\\
v_3
\end{bmatrix}$$

```octave
v = [1;2;3];
v = transpose([1 2 3]);
v = [1 2 3]'
```

## Array Operations

Addition, subtraction and scalar multiplication are defined by the corresponding array operators, eg.

```octave
w = [-3 3 -1]';
(v+w)'
(-w)'
(v-w)'
(4*(v-w))'
```

## Length

The length of a vector in $\mathbb R^3$ is given by
$$||\bar v|| = \sqrt{v_1^2+v_2^2+v_3^2}$$
The matlab command to calculate the length is `norm`:

```octave
norm(v)
sqrt(v(1)^2+v(2)^2+v(3)^2)
```

Do not use the command `length` that gives the number of elements in the longest dimension:

```octave
length(v)
```

## Vector operations

### Dot product

The *dot* product between the vectors $\bar v$ and $\bar w$ is defined by
$$\bar v \cdot \bar w = v_1w_1+v_2w_2+v_3w_3$$
Matlab gives the command `dot` to evaluate the *inner* or *scalar product*:

```octave
dot([1;2;3],[1;1;1])
dot(v,v)
norm(v)^2
```

The angle between two vectors is given by
$$\cos\theta=\frac{\bar v\cdot\bar w}{||\bar v||\,||\bar w||}$$

```octave
cos_theta = dot(v,w)/norm(v)/norm(w)
```

$\bar v$ and $\bar w$ are *orthogonal*.

### Cross product

The *cross* product between the vectors $\bar v$ and $\bar w$ is defined by
$$\bar v \times \bar w = \begin{vmatrix}
\bar e_1& \bar e_2& \bar e_3\\
v_1&v_2&v_3\\
w_1&w_2&w_3
\end{vmatrix} = \begin{bmatrix}
v_2w_3-v_3w_2\\
v_3w_1-v_1w_3\\
v_1w_2-v_2w_1
\end{bmatrix}$$

```octave
u = cross(v,w)
```

```octave
dot(v,u)
dot(w,u)
```

The *cross product* of $\bar v$ and $\bar w$ is a vector perpendicular to $\bar v$ and $\bar w$, its norm is given by
$$||\bar v\times\bar w|| = \sin\theta\,||\bar v||\,||\bar w||$$

where $\theta$ is the oriented angle between $\bar v$ and $\bar w$ and it points in the direction specified by the right hand rule.

```octave
disp(cross([1;0;0],[0;1;0])')
disp(cross([0;1;0],[0;0;1])')
disp(cross([0;0;1],[1;0;0])')
```

## Plotting in 3D

The following example plots a series of contours, meshes and surfaces for $z=25-x^2-y^2$ with $x\in[-5,5]$ and $y\in[-5,5]$.

```octave
%plot --size 1200,400 --format svg
[x, y] = meshgrid(-5:0.5:5,-5:0.5:5);
z = 25-x.^2-y.^2;
subplot(1,3,1); contour(x, y, z);
subplot(1,3,2); mesh(x, y, z);
subplot(1,3,3); surf(x, y, z)
```

The following commands plot a spiral in 3D

```octave
%plot --size 600,400 --format svg
t = 0:0.1:6*pi;
x = cos(t);
y = sin(t);
z = t/20;
plot3(x,y,z,'linewidth',3);
axis square;
```

# Matrices

## Definition

A matrix is an array with 2 dimensions: i.e. the rows and the columns, eg.

```octave
A = magic(3);
```

## Array Operations

All the operations defined for an array can be executed for a matrix:

```octave
A + A' % a symmetric matrix
3*A
A*A
```

## Matrix product

The matrix product of $\mathbf A$ with size $m \times r$ and $\mathbf B$ with size $r\times n$, noted $\mathbf{AB}$, is a matrix $\mathbf C$ with size $m\times n$ where
$$\mathbf C_{i,j}=\mathbf A_{i,1}\mathbf B_{1,j}+\mathbf A_{i,2}\mathbf B_{2,j}+\cdots+\mathbf A_{i,r}\mathbf B_{r,j}$$
or $\mathbf C_{i,j}$ is the dot product of row $i$ of $\mathbf A$ and column $j$ of $\mathbf B$.

In general $\mathbf{AB}\neq\mathbf{BA}$, that is, matrices are *noncommutative*.

```octave
A = [1 2
     3 4];
B = [1 -1
     1  2];
A*B
B*A
```

## Determinant

A determinant of a square matrix can be defined in a recursive way. Let $\mathbf A$ a $2\times2$ matrix:
$$\mathbf A = \begin{bmatrix}
a& b\\c&d\end{bmatrix}$$
then
$$\det(\mathbf A)=ad-bc$$
The determinant of a $3 \times 3$ matrix $\mathbf B=\begin{bmatrix}
b_{1,1}&b_{1,2}&b_{1,3}\\
b_{2,1}&b_{2,2}&b_{2,3}\\
b_{3,1}&b_{3,2}&b_{3,3}\end{bmatrix}$ is
$$\det(\mathbf B)=b_{1,1}\begin{vmatrix}
b_{2,2}&b_{2,3}\\
b_{3,2}&b_{3,3}\end{vmatrix}
-b_{1,2}\begin{vmatrix}
b_{2,1}&b_{2,3}\\
b_{3,1}&b_{3,3}\end{vmatrix}
+b_{1,3}\begin{vmatrix}
b_{2,1}&b_{2,2}\\
b_{3,1}&b_{3,2}\end{vmatrix}$$
This procedure is called cofactor expansion and can be done along each column or row. Attention to the sign before the subdeterminants!

```octave
det(A)
B = magic(3)
det(B)
```

## Inverse

A square matrix can have an inverse defined by
$$\mathbf A\mathbf A^{-1} = \mathbf A^{-1}\mathbf A = \mathbf{I}$$
where $\mathbf{I}$ is the unit matrix. A matrix is not invertible if its determinant equals zero.

```octave
inv(A)
B = [ 1  2
     -1 -2]
det(B)
inv(B)
```

## Linear systems

A linear system
$$\begin{cases}
a_{1,1}x_1+a_{1,2}x_2+\cdots+a_{1,n}x_n=b_1\\
a_{2,1}x_1+a_{2,2}x_2+\cdots+a_{2,n}x_n=b_2\\
\vdots\\
a_{n,1}x_1+a_{n,2}x_2+\cdots+a_{n,n}x_n=b_n\\
\end{cases}$$
can be represented as a matrix-vector equation:
$$\mathbf A\bar x=\bar b$$
and its solution is
$$\bar x=\mathbf A^{-1}\bar b$$

```octave
A = [1 1  2
     2 4 -3
     3 6 -5];
b = [1; 5; 2];
x = inv(A)*b
```

Another way to solve this system is based on Gaussian elimination called the *row reduced echelon form*, eg. 

```octave
rref([A, b])
```

Matlab has an operator that uses an optimized (numerical) method to solve a linear system:

```octave
x = A\b % or linsolve(A, b)
```

## Eigenvalue and Eigenvectors

If $\mathbf A$ is a square matrix then a scalar $\lambda$ is called an *eigenvalue* of $\mathbf A$, if associated with it there is a non-zero vector $\bar v$, called an *eigenvector*, such that
$$\mathbf A\bar v=\lambda \bar v$$

The eigenvalues can be find solving the characteristic polynomial:
$$\det(\mathbf A-\lambda\mathbf I)=0$$

The eigenvalues can then be found:
$$(\mathbf A-\lambda\mathbf I)\bar v=0$$

```octave
A = [0 1
     1 0];
p = poly(A) % characteristic polynomial
lambdas = roots(p)'
lambdas = eigs(A)'
[P, D] = eig(A)
```

## Orthogonal diagonalisation

A $n \times n$ matrix $\mathbf A$ is *diagonalisable* if one can find an invertible matrix $\mathbf P$, such that
$$\mathbf P^{-1}\mathbf A\mathbf P=\mathbf D$$
or
$$\mathbf A\mathbf P=\mathbf P\mathbf D\mathbf P^{-1}$$
where $\mathbf D$ is diagonal.

```octave
A = [3 1
     0 2];
[P, D] = eig(A)
DD = inv(P)*A*P
AA = P*D*inv(P)
```

If the matrix $\mathbf A$ is real and symmetric then:
1. the eigenvectors are orthogonal;
2. if the eigenvectors are normalised then $\mathbf P^{-1}=\mathbf P^T$;
3. the eigenvalues will be real.

```octave
A = [9 2
     2 6];
[P, D] = eig(A)
inv(P)
```

# Linear independence

A set of vectors is linearly independent if the determinant of the matrix with the vectors as columns is not zero.

```octave
v1 = [1;1;0];
v2 = [0;2;1];
v3 = [0;1;0];
det([v1,v2,v3]) % independent
v1 = [1;1;0];
v2 = [2;1;0];
v3 = [1;0;0];
det([v1,v2,v3]) % dependent
```
